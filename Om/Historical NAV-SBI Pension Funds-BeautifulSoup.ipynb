```python
PATH_LINK_CD = "/Users/om/opt/anaconda3/bin/chromedriver" #defined path link
driver = webdriver.Chrome(PATH_LINK_CD)

driver.get("https://www.sbipensionfunds.com/historical-nav/")

DateFrom = driver.find_element_by_name("fromdate")#Date Selection
DateFrom.send_keys("15-05-2009")


DateTo = driver.find_element_by_name("todate")
DateTo.send_keys(date.today().strftime("%d-%m-%Y"))

submit_button_click = driver.find_element_by_name("mysubmit")
submit_button_click.click()
                 
sleep(3)
```

```python
web_site_ps = driver.page_source#Get page source to python
```

```python
soup = BeautifulSoup(web_site_ps, 'html.parser')#parsing data through bs
```

```python
soup.title.text
```

```python
soup.prettify
```

```python
table = soup.find('table', attrs={'class':'table table-hover table-condensed table-bordered'})
trs = table.findAll("tr")
```

```python
final_table = [] #appending data in final_table
for tr in trs:
    td = tr.findAll("td")
    r = [tr.text.strip() for tr in td if tr.text.strip()]
    final_table.append([tr.text.strip() for tr in td if tr.text.strip()])
    
    if r:
        final_table.append(r)
```

```python
data_frame = pd.DataFrame(final_table)

print(data_frame)

data_frame.to_csv('Historical_NAV_csvfile2', index=False)
#data_frame = pd.read_csv('Historical_NAV_csvfile.csv1', index_col=0)

sleep(9)
```
